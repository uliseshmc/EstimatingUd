```python
folder_in = paths.DATA_HUMCHIMPORANG115 + 'cds/chrm22/'
#folder_in = 'cds/codon_aligned/'

sequence = 'ENSG00000073169.fa'

aln = load_unaligned_seqs(filename = folder_in + sequence, moltype='dna')
rename = libs.renamer_cds_unaligned()
trim_stops = get_app("trim_stop_codons")

aln = rename(aln)
aln = trim_stops(aln)
aln


#folder_in + sequence
```

```python
codon_align = get_app("progressive_align", "codon", guide_tree="(Human:0.06,Chimpanzee:0.06,Orangutan:0.14)")

aln_codon = codon_align(aln)
aln_codon
```

```python
stop_codons = {"TAA", "TAG", "TGA"}

def find_internal_stop(seq):
    dna = str(seq)
    # if you expect a terminal stop codon, exclude last codon:
    for i in range(0, len(dna), 3):
        codon = dna[i:i+3]
        if codon in stop_codons:
            return True, i, codon
    return False, None, None
```

```python
for name in aln.names:
    seq = aln.seqs[name]
    has_stop, pos, codon = find_internal_stop(seq)
    if has_stop:
        print(name, "has internal stop codon", codon, "at nt position", pos + 1)
```
