```python
import cogent3
from cogent3 import get_app
from cogent3 import load_aligned_seqs
from cogent3 import load_unaligned_seqs
import paths
import libs
```

```python
folder_in =  paths.DATA_APES114 + 'introns/'

sequence = 'ENSG00000100033.fa'

aln = load_aligned_seqs(filename = folder_in + sequence, format_name="fasta", moltype='dna')

rename_noncds = libs.renamer_noncds_aligned()
aln = rename_noncds(aln)
```

```python
#sequences that use this app need to be renamed before
@cogent3.app.composable.define_app
def sample_UTR3(aln: cogent3.app.typing.AlignedSeqsType) -> cogent3.app.typing.AlignedSeqsType:
    seq = aln.get_gapped_seq("Human")

    try:
        pos3 = str(seq).rindex("?")
    except ValueError:
        pos3 = None

    UTR3_aln = aln[pos3+1:]

    return UTR3_aln

@cogent3.app.composable.define_app
def sample_UTR5(aln: cogent3.app.typing.AlignedSeqsType) -> cogent3.app.typing.AlignedSeqsType:
    seq = aln.get_gapped_seq("Human")

    try:
        pos5 = str(seq).index("?")
    except ValueError:
        pos5 = None

    UTR5_aln = aln[0:pos5]

    return UTR5_aln

@cogent3.app.composable.define_app
def removeUTRs_fromintrons(aln: cogent3.app.typing.AlignedSeqsType) -> cogent3.app.typing.AlignedSeqsType:
    seq = aln.get_gapped_seq("Human")

    try:
        pos5 = str(seq).index("?")
    except ValueError:
        pos5 = None

    try:
        pos3 = str(seq).rindex("?")
    except ValueError:
        pos3 = None

    intron_noUTR = aln[pos5:pos3]

    return intron_noUTR
```

```python
get_UTR3 = sample_UTR3()
aln_UTR3 = get_UTR3(aln)
aln_UTR3.set_repr_policy(num_pos = 400)
aln_UTR3
```

```python
get_UTR5 = sample_UTR5()
aln_UTR5 = get_UTR5(aln)
aln_UTR5.set_repr_policy(num_pos = 400)
aln_UTR5
```

```python
remove_UTR = removeUTRs_fromintrons()
intron_noUTR = remove_UTR(aln)
intron_noUTR.set_repr_policy(num_pos = 400)
intron_noUTR
```
