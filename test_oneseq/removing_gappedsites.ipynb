```python
import cogent3
from cogent3 import get_app
from cogent3 import load_aligned_seqs
from cogent3 import load_unaligned_seqs
import paths
import libs
```

```python
#Cleaning sequences first (omit_gap_pos_app)
folder_in = 'introns/'

sequence = 'ENSG00000100373.fa'

aln = load_aligned_seqs(filename = folder_in + sequence, moltype='dna')
name_map = {
        "homo_sapiens": "Human",
        "pan_troglodytes": "Chimpanzee",
        "gorilla_gorilla": "Gorilla"
    }

aln = aln.renamed_seqs(lambda x: name_map.get(x.split(":")[0], x))
omit_gap_pos_app = get_app("omit_gap_pos", moltype="dna")
aln = omit_gap_pos_app(aln)
omit_degs_cds = get_app("omit_degenerates", moltype="dna", motif_length=3)
aln = omit_degs_cds(aln)
aln
```
