```python
import cogent3
from cogent3 import get_app
from cogent3 import load_aligned_seqs

import trinuc_models as trinucs # this module must be in the same directory as this notebook
```

```python
sm=trinucs.GT_CpG_ss()
```

```python
paramnames = sm.get_param_list()
rules = [{"par_name": n, "is_independent": True} for n in paramnames if n!="omega"] + [{"par_name": "omega", "value": 1.0, "is_constant": True}]
rules
```

```python
folder_in = 'igar/'

sequence = 'homo_sapiens-22-36038797-36056389.fa'

just_nucs = get_app("omit_degenerates", moltype="dna")

aln_igar = load_aligned_seqs(filename = folder_in + sequence, moltype='dna')
aln_igar = just_nucs(aln_igar)
```

defining some optimisation args to prevent full optimisation so I can try things out fast without concer about accuracy of estimates

```python
opt_args = {"max_evaluations": 1000, "limit_action": "ignore"}
```

as this alignment is not a CDS, so we can only use the trinuc model. Note I set omega here to 1.0 since we know there's no natural selection for protein coding content on non-CDS sequences and we don't want to confound CpG parameter estimation.

```python
mod = cogent3.get_app("model", "GT_CpG_ss",
                      show_progress=True,
                      optimise_motif_probs=False,
                      #opt_args=opt_args, # uncomment this line to do "quick" experiments
                      param_rules=rules)
```

```python
result1 = mod(aln_igar)
```

```python
brca1 = cogent3.get_dataset("brca1")
new_aln = brca1.take_seqs(aln_igar.names).omit_gap_pos(motif_length=3, allowed_gap_frac=0)
```

```python
new_lf = trinucs.modified_lf(result1.lf)
new_lf.set_alignment(new_aln, motif_pseudocount=1e-12)
```
