```python
import cogent3
from cogent3 import get_app
from cogent3.maths.matrix_exponential_integration import expected_number_subs
import matplotlib.pyplot as plt
import paths
import libs
import pandas as pd
import numpy as np

chromosomes = list(range(1, 23))
```

## Intergenic Ancestral Repeats (IGAR)

```python
total_length_IGAR=0

for chrom_coord in chromosomes:
    folder_in = paths.DATA_APES114 + 'intergenicAR/chrm' + str(chrom_coord) + '/'
    in_dstore = cogent3.open_data_store(folder_in, suffix='fa', mode='r')
    
    #loading and cocatanating sequences
    #rename_noncds eliminates files that do not contain Gorilla, Human and Chimp sequences or that has duplicates.
    loader = get_app("load_unaligned", moltype="dna")
    seq_length = libs.human_seq_length_nonfiltered_noncds()
    
    totallength_app = loader + seq_length
    
    seq_lengths = [r for r in totallength_app.as_completed(in_dstore[:], parallel=True) if r]
    
    values = [x.obj for x in seq_lengths]
    temp_length = sum(values)
    total_length_IGAR = total_length_IGAR + temp_length
total_length_IGAR
```

## CDS

```python
total_length_cds=0

for chrom_coord in chromosomes:
    folder_in = paths.DATA_APES114 + 'cds/chrm' + str(chrom_coord)
    in_dstore = cogent3.open_data_store(folder_in, suffix='fa', mode='r')
    
    loader_cds = get_app("load_unaligned", moltype="dna")
    seq_length_cds = libs.human_seq_length_nonfiltered_cds()
    
    totallength_cds_app = loader_cds + seq_length_cds
    
    seq_lengths = [r for r in totallength_cds_app.as_completed(in_dstore[:], parallel=True) if r]
    
    values = [x.obj for x in seq_lengths]
    temp_length = sum(values)
    total_length_cds = total_length_cds + temp_length

total_length_cds
```

## Introns

```python
total_length_introns=0
for chrom_coord in chromosomes:
    folder_in = paths.DATA_APES114 + 'introns/chrm' + str(chrom_coord)
    in_dstore = cogent3.open_data_store(folder_in, suffix='fa', mode='r')
    
    seq_lengths = [r for r in totallength_app.as_completed(in_dstore[:], parallel=True) if r]
    
    values = [x.obj for x in seq_lengths]
    temp_length = sum(values)
    total_length_introns = total_length_introns + temp_length
```
