# Calculate number of motifs per region

## Intergenic ancestral repeat

```python
folder_in = paths.DATA_APES114 + 'intergenicAR/'
in_dstore = cogent3.open_data_store(folder_in, suffix='fa', mode='r')

#loading and cocatanating sequences
#rename_noncds eliminates files that do not contain Gorilla, Human and Chimp sequences or that has duplicates.
loader = get_app("load_aligned", moltype="dna")
omit_degs_noncds = get_app("omit_degenerates", moltype="dna")
rename_noncds = libs.renamer_noncds_aligned()
table_motifs = libs.number_of_motifs()

noncds_app = loader + omit_degs_noncds + rename_noncds + table_motifs

motifs = [r for r in noncds_app.as_completed(in_dstore[:], parallel=True) if r]
#net line takes the value from the motifs cogent3.app.composable.source_proxy variable
motifs_Counter = [x.obj for x in motifs]
motifs_IGAR = sum(motifs_Counter, Counter())
#motifs_IGAR
```

```python
folder_in = paths.DATA_APES114 + 'cds/codon_aligned'
in_dstore = cogent3.open_data_store(folder_in, suffix='fa', mode='r')

loader_cds = get_app("load_aligned", moltype="dna")   
omit_degs_cds = get_app("omit_degenerates", moltype="dna", motif_length=3)
table_motifs = libs.number_of_motifs()

cds_process = loader_cds + omit_degs_cds + table_motifs

motifs = [r for r in cds_process.as_completed(in_dstore[:], parallel=True) if r]

motifs_Counter = [x.obj for x in motifs]
motifs_cds = sum(motifs_Counter, Counter())
```

Calculate the mutation rate constant constant_mu

```python
Human_mutation_rate = 1.34*10**(-8)

sitesTimesPropMutation = Counter(
    {k: motifs_totals[k] * Prop_model[k] for k in motifs_totals.keys() & Prop_model.keys()}
)

sumSitesTimesPropMutation = sum(sitesTimesPropMutation.values())

constant_mu = Human_mutation_rate*Total_sites/sumSitesTimesPropMutation
with open("Output_data/constantofmu.txt", "w") as f:
    f.write(str(constant_mu))
```

# Create data frame of Number of sites times relative mutation rate

```python
#CDS
sitesTimesPropMutation_cds = Counter(
    {k: motifs_cds[k] * Prop_model[k] for k in motifs_cds.keys() & Prop_model.keys()}
)
sumSitesPropMut_cds = sum(sitesTimesPropMutation_cds.values())

#introns
sitesTimesPropMutation_introns = Counter(
    {k: motifs_introns[k] * Prop_model[k] for k in motifs_introns.keys() & Prop_model.keys()}
)
sumSitesPropMut_introns = sum(sitesTimesPropMutation_introns.values())

#5UTR
sitesTimesPropMutation_5UTR = Counter(
    {k: motifs_5UTR[k] * Prop_model[k] for k in motifs_5UTR.keys() & Prop_model.keys()}
)
sumSitesPropMut_5UTR = sum(sitesTimesPropMutation_5UTR.values())

#3UTR
sitesTimesPropMutation_3UTR = Counter(
    {k: motifs_3UTR[k] * Prop_model[k] for k in motifs_3UTR.keys() & Prop_model.keys()}
)
sumSitesPropMut_3UTR = sum(sitesTimesPropMutation_3UTR.values())

#intronAR
sitesTimesPropMutation_intronAR= Counter(
    {k: motifs_intronAR[k] * Prop_model[k] for k in motifs_intronAR.keys() & Prop_model.keys()}
)
sumSitesPropMut_intronAR = sum(sitesTimesPropMutation_intronAR.values())

#distalIG
sitesTimesPropMutation_distalIG= Counter(
    {k: motifs_distalIG[k] * Prop_model[k] for k in motifs_distalIG.keys() & Prop_model.keys()}
)
sumSitesPropMut_distalIG = sum(sitesTimesPropMutation_distalIG.values())

#proximal5IG
sitesTimesPropMutation_proximal5IG= Counter(
    {k: motifs_proximal5IG[k] * Prop_model[k] for k in motifs_proximal5IG.keys() & Prop_model.keys()}
)
sumSitesPropMut_proximal5IG = sum(sitesTimesPropMutation_proximal5IG.values())

#_intronAR
sitesTimesPropMutation_proximal3IG= Counter(
    {k: motifs_proximal3IG[k] * Prop_model[k] for k in motifs_proximal3IG.keys() & Prop_model.keys()}
)
sumSitesPropMut_proximal3IG = sum(sitesTimesPropMutation_proximal3IG.values())
```

```python
sumSitesPropMut = pd.DataFrame([
        ['cds', sumSitesPropMut_cds],
        ['introns', sumSitesPropMut_introns],
        ['3UTR', sumSitesPropMut_3UTR],
        ['5UTR', sumSitesPropMut_5UTR],
        ['intronsAR', sumSitesPropMut_intronAR],
        ['distalIG', sumSitesPropMut_distalIG],
        ['proximal5IG', sumSitesPropMut_proximal5IG],
        ['proximal3IG', sumSitesPropMut_proximal3IG],
    ],
    columns = ['Sequence type', 'Sum NumbSites PropMut'])

sumSitesPropMut.plot(
    x='Sequence type',
    kind = 'bar',
    rot=45)
```
