```python
import cogent3
from cogent3 import get_app
from cogent3 import load_aligned_seqs
import matplotlib.pyplot as plt
import libs
import paths
```

```python
#Define apps
omit_degs_noncds = get_app("omit_degenerates", moltype="dna")
rename_noncds = libs.renamer_noncds_aligned()
sm_GN = get_app("model", "GN", time_het="max", optimise_motif_probs=True, show_progress=True)
```

# Introns

```python
folder_in = paths.DATA_APES114 + 'introns/'

sequence = 'ENSG00000100373.fa'
aln1 = load_aligned_seqs(filename = folder_in + sequence, format_name="fasta", moltype='dna')
aln1 = rename_noncds(aln1)
aln1 = omit_degs_noncds(aln1)

result_introns1 = sm_GN(aln1)
lengthstree_introns1 = result_introns1.lf.get_lengths_as_ens()
#print("Model results : \n" + str(result_introns1.lf) + "\n")
print("Distances: " + str(lengthstree_introns1))
```

```python
folder_in = paths.DATA_APES114 + 'introns/nt_aligned_omitgaps/'

aln_ntaligned1 = load_aligned_seqs(filename = folder_in + sequence, format="fasta", moltype='dna')
aln_ntaligned1 = omit_degs_noncds(aln_ntaligned1)


result_introns_ntaligned1 = sm_GN(aln_ntaligned1)
lengthstree_introns_ntaligned1 = result_introns_ntaligned1.lf.get_lengths_as_ens()
#print("Model results : \n" + str(result_introns_ntaligned1.lf) + "\n")
print("Distances: " + str(lengthstree_introns_ntaligned1))
```
